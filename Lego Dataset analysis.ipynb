```python
#Jointure 

sets_themes = pd.merge(sets, themes, how = 'left', left_on = 'theme_id', right_on = 'id', suffixes = ('_sets', '_themes'))
sets_themes.head()
```

```python
sets_themes.columns
```

```python
#nombre total de num_parts par themes
num_parts_per_theme = sets_themes.groupby('name_themes').sum().reset_index()
num_parts_per_theme = num_parts_per_theme[['name_themes', 'num_parts']].sort_values('num_parts', ascending = False)
num_parts_per_theme
```

```python
#nombre total de num_parts par année
num_parts_per_theme = sets_themes.groupby('year').sum().reset_index()
num_parts_per_theme = num_parts_per_theme[['year', 'num_parts']].sort_values('num_parts', ascending = False)
num_parts_per_theme
```
