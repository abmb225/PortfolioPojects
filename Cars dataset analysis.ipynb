```python
import pandas as pd
file_path = r"C:\Users\fredd\Downloads\file (2).csv"
data = pd.read_csv(file_path)
data.head()
```

```python
#Trouver les valeurs nulles par colonnes et les remplacer par la moyenne de celle-ci
data.isnull().sum()
```

```python
#regrouper les colonnes en int 
columns_to_fill = ['EngineSize', 'Cylinders', 'Horsepower', 'MPG_City', 'MPG_Highway', 'Weight', 'Wheelbase', 'Length']

#boucle for pour remplacer les valeurs nulles par la moyenne de la colonne
for column in columns_to_fill : 
    data[column].fillna(data[column].mean(), inplace = True)

data.head(50)
```

```python
data['Horsepower'].mean()
```

```python
#observer les diffrents type de Make et leurs frquences
data['Make'].value_counts()
```

```python
#Montrer les enregistrements o Oringin = Asia / Europe

data[data['Origin'] .isin(['Asia', 'Europe'])]
data.head(100)
```

```python
#enlever des enregistrements non-dsirs (weight > 4000)
data[~(data['Weight'] > 4000)]
```

```python
#augmenter les valeurs de MPG_city de 3 
data['MPG_City'] = data['MPG_City'].apply(lambda x:x+3)
```
